# Read Total profit of all months and show it using a line plot  
X label name = Month Number  
Y label name = Total profit  

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

## write your code here
df = pd.read_csv('company_sales_data.csv')
plt.plot(df.month_number,df.total_profit)
plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.title("Company profit per month")
plt.xticks(np.linspace(1,12,12,endpoint=True))
plt.yticks(np.linspace(100000,500000,5,endpoint=True))
plt.show()
```

# Calculate total sale data for last year for each product and show it using a Pie chart  
Note: In Pie chart display Number of units sold per year for each product in percentage

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt  

## write your code here
df = pd.read_csv('company_sales_data.csv')
df = df.drop(["month_number","total_units","total_profit"], axis=1)

labels = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
colors = ['blue', 'gold', 'green', 'red', 'purple', 'brown']
total = df.apply(sum).sum()
sizes = []
for i in df[labels]:
    sizes.append(df[i].sum()/total*100)
fig1, ax1 = plt.subplots()
ax1.pie(sizes,colors=colors, startangle=0,labels=labels, autopct='%1.1f%%')
ax1.legend(labels, loc=4)
ax1.axis('equal')
plt.title("Sales data")
plt.show()
```
